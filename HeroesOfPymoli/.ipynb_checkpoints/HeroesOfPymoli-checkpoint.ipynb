```python
import pandas as pd 
```

```python
purchase_file = "../purchase_data.csv"
```

```python
purchase_data = pd.read_csv(purchase_file)
purchase_data.head()
```

```python
players = len(purchase_data["SN"].unique())

players_count = pd.DataFrame({"Total Players": [players]})

players_count
```

```python
# calculations to obtain number of unique items, average price, etc

unique_items = len(purchase_data["Item Name"].unique())
average_price = purchase_data["Price"].mean()
number_of_purchases = len(purchase_data["Purchase ID"])
total_revenue = purchase_data["Price"].sum()

#Create a summary data frame to hold the results

purchase_data_df = pd.DataFrame({"Number of Unique Items": [unique_items],
                                "Average Price": [average_price],
                                "Number of Purchases": [number_of_purchases],
                                "Total Revenue": [total_revenue]})

#Optional: give the displayed data cleaner formatting

purchase_data_df["Average Price"] = purchase_data_df["Average Price"].map("${:,.2f}".format)
purchase_data_df["Total Revenue"] = purchase_data_df["Total Revenue"].map("${:,.2f}".format)

purchase_data_df


```

```python
#Percentage and Count of Male Players
#Percentage and Count of Female Players
#Percentage and Count of Other / Non-Disclosed


gender_demo_df = purchase_data["Gender"].value_counts()
gender_demographics_df = pd.DataFrame(gender_demo_df)

gender_percent = (purchase_data["Gender"].value_counts()/ (purchase_data["Gender"].count()))*100
gender_demographics_df["Percent of Players"] = gender_percent.map("{:,.2f}%".format)
Gender_df = pd.DataFrame(gender_demographics_df)

gender_DF = Gender_df.rename(columns={"Gender": "Total Count"})
gender_DF






```

```python
genders_count_df = purchase_data.groupby("Gender")

#sn_count_sum = genders_count_df["SN"].sum()
sn_count = genders_count_df["SN"].count()
sn_count_df = pd.DataFrame(sn_count)


sn_count_df.dtypes
```

```python
average_price = genders_count_df["Price"].mean()
total_purchase = genders_count_df["Price"].sum()

#sn_sum = 780

avg_total_person = total_purchase / sn_count

sn_summary_df = pd.DataFrame({'Purchase Count': sn_count, 
                             'Average Purchase Price': average_price.map("{:,.2f}%".format),
                             'Total Purchase Value': total_purchase,
                             'Average Total Puchase': avg_total_person.map("{:,.2f}%".format)})

sn_summary_df
```

```python
bins = [0,10,15,20,25,30,35,40,100]
group_names=["<10","11-15","16-20","21-25","26-30","31-35","35-40","40+"]

purchase_data["Total Count"] = pd.cut(purchase_data["Age"], bins, labels=group_names)
groupby_age = purchase_data.groupby("Total Count")
#groupby_sum = purchase_data["Total Count"].count()
#percentage_of_players = purchase_data["Total Count"]/ (groupby_sum)
#purchase_data["Percentage of Players"] = percentage_of_players
groupby_age_df = groupby_age.count()

age_summary_table = groupby_age_df[['Age']]
age_summary_table
```

```python
bins = [0,10,15,20,25,30,35,40,100]
group_names=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

purchase_analysis = purchase_data.groupby('Total Count')['Price'].count()
purchase_price_df = pd.DataFrame(purchase_analysis)

price_sum_df = purchase_data.groupby('Total Count')['Price'].sum()

avg_price = purchase_data.groupby('Total Count')['Price'].mean()
purchase_price_df['Average Purchase Price'] = avg_price.map("${:,.2f}".format)

purchase_price_df['Total Purchase Value'] = price_sum_df.map("${:,.2f}".format)

renamed_price_df = purchase_price_df.rename(columns={'Price': 'Purchase Count'})

descending_price_df = renamed_price_df.sort_values(by=['Total Purchase Value'], ascending=False)
descending_price_df.head()

#purchase_data["Total Count"] = pd.cut(purchase_data["Age"], bins, labels=group_names)
#groupby_age = purchase_data.groupby("Total Count")
#groupby_age[["Purchase ID"]].count()
```

```python
count_sn = purchase_data.groupby('SN')['Price'].count()
count_sn_df = pd.DataFrame(count_sn)

SN_sum_df = purchase_data.groupby('SN')['Price'].sum()

avg_price_sn = purchase_data.groupby('SN')['Price'].mean()
count_sn_df['Average Purchase Price'] = avg_price_sn.map("${:,.2f}".format)

count_sn_df['Total Purchase Value'] = SN_sum_df.map("${:,.2f}".format)

renamed_count_sn_df = count_sn_df.rename(columns={'Price': 'Purchase Count'})

descending_sn_df = renamed_count_sn_df.sort_values(by=['Total Purchase Value'], ascending=False)
descending_sn_df.head()


#renamed_count_sn_df
```

```python
item_id = purchase_data.groupby(purchase_data['Item ID'])
unique_items = item_id['Item ID'].unique().str[0]
item_name = item_id['Item Name'].unique().str[0]
age_purchase = item_id['Age'].count()
item_price = item_id['Price'].unique().str[0]
purchase_total = item_id['Price'].sum()

items_summary = pd.DataFrame({'Item ID': unique_items,
                             'Item Name': item_name,
                             'Item Price': item_price.map("${:,.2f}".format),
                             'Item Count': age_purchase,
                             'Total Purchase': purchase_total.map("${:,.2f}".format)})
items_summary = items_summary.sort_values('Item Count', ascending=False)
items_summary_df = items_summary[['Item Name', 'Item Count', 'Item Price','Total Purchase']]
items_summary_df.head()
```
